```python
# Importing the dependecies 
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Load dataset to dataframe
data = pd.read_csv("creditcard.csv")
data.head()
```

```python
# information related to the dataset
data.info()
```

```python
# Check the number of null values in each column 
data.isnull().sum()
```

```python
# Distribution of legit and fraud transactions in the dataset
data['Class'].value_counts()
```

```python
''' This is a unbalanced dataset as the o values are comparitively more than 1
labels 
0 == Good Transaction 
1 == Fraud Transaction '''
```

```python
# Separate the data for analysis 
good = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
print(good.shape)
print(fraud.shape)
```

```python
# Statistical measure of the data

good.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values of both goood and fraud

data.groupby('Class').mean()
```

```python
'''
Under-Sampling :
Build a sample dataset containing simillar distribution containing good and fraud transactions
Number or fraud transaction = 492 & good transaction = 284315
'''
```

```python
good_sample = good.sample(n=492)
```

```python
# Concatenating 2 dataframs good_sample and fraud

new_data = pd.concat([good_sample,fraud],axis = 0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
data.groupby('Class').mean()
```

```python
# Splitting the data into features and targets

X = new_data.drop(columns = 'Class',axis = 1)
Y = new_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Split the data into training and testing data 

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,stratify = Y, random_state = 2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
'''
Model Training : Logistic Regressioin 
'''

model = LogisticRegression()
```

```python
# Training the LogisticRegression with training data

model.fit(X_train,Y_train)
```

```python
# Evaluation of the Model based on accuracy score

X_train_prediction = model.predict(X_train)
train_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on the training data : ",train_data_accuracy)
```

```python
# Accuracy on test data

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on the test data : ",test_data_accuracy)
```
